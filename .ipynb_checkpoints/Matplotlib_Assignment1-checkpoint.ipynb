```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
#resource of data is https://www.w3schools.com/python/pandas/pandas_dataframes.asp
MyData={'Duration':[60,60,60,60,45,45,45,45,30,30,30,30],
        'Burning calories':[409.1,479,340,269,282.4,246,406,253.3,240.9,250.4,195.1,238]}
df = pd.DataFrame(MyData)

plt.plot(df, linestyle = 'dotted')
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python

total_Burning_calories = df['Burning calories'].sum()
piedata=df.groupby('Duration',as_index=False).sum()
piedata2=piedata.replace(to_replace = 45, value = 'Three quarters')
piedata2=piedata2.replace(to_replace = 30, value = 'Half hour')
piedata2=piedata2.replace(to_replace = 60, value = '1 Hour')
plt.pie(piedata['Burning calories'],labels=piedata2['Duration'], autopct='%1.2f%%', shadow=True, radius=1.5, explode = [0.3,0,0])
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(df['Duration'],df['Burning calories'], marker = 'o',alpha = 0.4,s= 50) 
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(df['Duration'],df['Burning calories'],alpha = 0.4,s= 50)
plt.xlabel('Duration',fontsize = 16)
plt.ylabel('Burning calories',fontsize = 16)
plt.title('The number of calories burned during workout based on activity duration', color = 'red',loc = 'left')
```

## Q5: Create three line charts with Legends at left and colors.

```python
Data={'Duration':[60,60,60,60,45,45,45,45,30,30,30,30],
        'Burning calories':[409.1,479,340,269,282.4,246,406,253.3,240.9,250.4,195.1,238],
      'Puls':[108,106,102,98,105,100,99,104,108,107,90,101] }
df1 = pd.DataFrame(Data)

plt.plot(df1['Duration'], 'g')
plt.plot(df1['Burning calories'], 'b')
plt.plot(df1['Puls'], 'r')
plt.legend(['Duration','Burning calories','Puls'], shadow = True, loc = 'upper left')
plt.show()
```
